```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("./Data Preprocessing/ProcessedData/Delhi/pusa,-delhi-air-quality.csv")


df['date'] = pd.to_datetime(df['date'])


start_date = "2020-03-01"
end_date = "2022-12-31"


covid_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]


plt.figure(figsize=(12, 6))
plt.plot(covid_df['date'], covid_df['AQI'], label="AQI", color="b")
plt.axvline(pd.to_datetime("2020-03-01"), color="r", linestyle="--", label="COVID Start")
plt.axvline(pd.to_datetime("2022-12-31"), color="g", linestyle="--", label="COVID End")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI Trends During COVID-19 Period (March 2020 - December 2022)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("./Data Preprocessing/ProcessedData/Delhi/pusa,-delhi-air-quality.csv")


df['date'] = pd.to_datetime(df['date'])


df['year'] = df['date'].dt.year

yearly_aqi = df.groupby('year')['AQI'].mean().reset_index()


plt.figure(figsize=(12, 6))
plt.plot(yearly_aqi['year'], yearly_aqi['AQI'], marker='o', linestyle='-', color='b', label="Average AQI")
plt.xlabel("Year")
plt.ylabel("Average AQI")
plt.title("Yearly AQI Trends")
plt.legend()
plt.grid()
plt.xticks(yearly_aqi['year'])
plt.show()
```
